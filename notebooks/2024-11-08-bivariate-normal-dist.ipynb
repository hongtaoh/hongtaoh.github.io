```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
mu1 = 1
std1 = 1.5
mu2 = 1.2
std2 = 2
size = 100000
data1 = np.random.normal(mu1, std1, size)
data2 = np.random.normal(mu2, std2, size)
```

```python
sns.histplot(data1, bins=30, kde=True, color='g')
plt.title('Histogram of Normal Distribution (Data1) with KDE')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data2, bins=30, kde=True, color='steelblue')
plt.title('Histogram of Normal Distribution (Data2) with KDE')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x = data1, y = data2, size = 0.2)
```

```python
import numpy as np
import pandas as pd
import altair as alt

# Parameters for the bivariate normal distribution
mean = [0, 0]  # Mean vector for the two variables
cov = [[1, 0.7], [0.7, 1]]  # Covariance matrix (controls correlation, here 0.8)

# Generate random samples
data = np.random.multivariate_normal(mean, cov, 1000)
df = pd.DataFrame(data, columns=['X', 'Y'])

# Create the scatter plot
scatter_plot = alt.Chart(df).mark_circle(size=60).encode(
    x='X:Q',
    y='Y:Q',
    tooltip=['X', 'Y']
).properties(
    title='Scatter Plot of Bivariate Normal Distribution',
    width=600,
    height=400
)

scatter_plot.show()
```
