```julia
using GLMakie
using Interact
using Distributions

# Create an interactive plot with sliders for controlling the parameters
@manipulate for mu1 in -5:0.1:5, std1 in 0.1:0.1:5,
                     mu2 in -5:0.1:5, std2 in 0.1:0.1:5,
                     ρ in -1:0.05:1

    # Define the mean vector and covariance matrix
    μ = [mu1, mu2]
    Σ = [std1^2 ρ*std1*std2; ρ*std1*std2 std2^2]

    # Generate random samples from the bivariate normal distribution
    dist = MvNormal(μ, Σ)
    samples = rand(dist, 1000)'

    # Create the scatter plot
    fig = Figure(size = (800, 600))  # Use `size` instead of `resolution`
    ax = Axis(fig[1, 1], title="Interactive Scatter Plot of Bivariate Normal Distribution",
              xlabel="X", ylabel="Y")
    GLMakie.scatter!(ax, samples[:, 1], samples[:, 2], color=:blue, markersize=4)  # Specify `GLMakie.scatter!`

    # Display the figure
    fig
end
```
