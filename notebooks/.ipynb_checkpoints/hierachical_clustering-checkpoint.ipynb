```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
import matplotlib.pyplot as plt

from matplotlib.pyplot import figure
```

```python
np.random.seed(1)
N = 50
matrix = np.random.rand(N,2)
# multiply by ten so the numbers are larger and easier to understand
# otherwise, all of them will be between 0 and 1
M = matrix * 10
M
```

```python
X = M[:, 0]
Y = M[:, 1]
```

```python
figure(figsize=(8, 8), dpi=100)
plt.scatter(X, Y)
# https://stackoverflow.com/a/42554070
for i in range(N):
    plt.text(X[i], Y[i], str(i))
```

## Euclidean distance

```python
def eu_distance(x,y):
    d_square = np.sum((x-y)**2)
    d = np.sqrt(d_square)
    return d
```

```python
eu_distance(M[27],M[28])
```

## Distance matrix

```python
dist_matrix = np.zeros((N, N))
dist_matrix
```

```python
for i in range(N):
    for j in range(N):
        if i < j:
            dist_matrix[i,j] = eu_distance(M[i], M[j])
        else:
            dist_matrix[i,j] = 10**10
```

```python
# initiate clusters
clusters = [[i] for i in range(N)]
clusters
```

```python
def single_linkage_dist(cluster1, cluster2):
    '''calculate the distance between two clusters using single linkage method
    '''
    dist_list = []
    for i in cluster1:
        for j in cluster2:
            if i < j:
                dist_list.append(dist_matrix[i,j])
            else:
                dist_list.append(dist_matrix[j,i])
    return min(dist_list)
```

```python
def complete_linkage_dist(cluster1, cluster2):
    '''calculate the distance between two clusters using complete linkage method
    '''
    dist_list = []
    for i in cluster1:
        for j in cluster2:
            if i < j:
                dist_list.append(dist_matrix[i,j])
            else:
                dist_list.append(dist_matrix[j,i])
    return max(dist_list)
```

```python
def average_linkage_dist(cluster1, cluster2):
    '''calculate the distance between two clusters using average linkage method
    '''
    dist_list = []
    for i in cluster1:
        for j in cluster2:
            if i < j:
                dist_list.append(dist_matrix[i,j])
            else:
                dist_list.append(dist_matrix[j,i])
    return np.mean(dist_list)
```
