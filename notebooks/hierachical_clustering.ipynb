```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
import matplotlib.pyplot as plt

from matplotlib.pyplot import figure
```

```python
np.random.seed(1)
N = 50
matrix = np.random.rand(N,2)
M = matrix * 10
M
```

```python
X = M[:, 0]
Y = M[:, 1]
```

## Euclidean distance

```python
def eu_distance(x,y):
    d_square = np.sum((x-y)**2)
    d = np.sqrt(d_square)
    return d
```

```python
eu_distance(M[27],M[28])
```

## Distance matrix

```python
dist_matrix = np.zeros((N, N))
dist_matrix
```

```python
for i in range(N):
    for j in range(N):
        if i >= j:
            dist_matrix[i,j] = 10**10
        else:
            dist_matrix[i,j] = eu_distance(M[i], M[j])
```

```python
df = pd.DataFrame(dist_matrix)
df
```

```python
def single_linkage_dist(cluster1, cluster2):
    dist_list = []
    for i in cluster1:
        for j in cluster2:
            if i < j:
                dist_list.append(dist_matrix[i,j])
            else:
                dist_list.append(dist_matrix[j,i])
    return min(dist_list)
```

```python
def complete_linkage_dist(cluster1, cluster2):
    dist_list = []
    for i in cluster1:
        for j in cluster2:
            if i < j:
                dist_list.append(dist_matrix[i,j])
            else:
                dist_list.append(dist_matrix[j,i])
    return max(dist_list)
```

```python
def average_linkage_dist(cluster1, cluster2):
    dist_list = []
    for i in cluster1:
        for j in cluster2:
            if i < j:
                dist_list.append(dist_matrix[i,j])
            else:
                dist_list.append(dist_matrix[j,i])
    return np.average(dist_list)
```
